```python
import nibabel as nib
import os
import numpy as np
import math
import matplotlib.pyplot as plt
```

## Find mean and std dev of every MRI sequence for analysis

```python
base_dir = '../reg_dataset/'
max_val=-math.inf
# print(type(am))
for ind in range(1,369):
    patient=f'BraTS20_Training_{ind:03d}'
    mri = nib.load(os.path.join(base_dir + '/{}/{}'.format(patient, patient + '_t1.nii.gz'))).get_fdata().T
    max_m=np.amax(mri)
    mean_mri=np.mean(mri)
    std_mri=np.std(mri)
    print(f'ind: {ind}--Max:{max_m}, Min:{np.amin(mri)} ,mean:{mean_mri}, STD:{std_mri}')
    if(max_m>max_val):
        max_val=max_m
```

```python
print(max_val)
```

## Visual verification of data standardization 

```python
patient=f'BraTS20_Training_{31:03d}'
mri = nib.load(os.path.join(base_dir + '/{}/{}'.format(patient, patient + '_t1.nii.gz'))).get_fdata()
m = np.mean(mri)
s = np.std(mri)
plt.figure()
plt.title('raw')
plt.imshow(mri[:, :, 123], cmap=plt.cm.bone)
normed=(mri-m)/s
plt.figure()
plt.title('normed')
plt.imshow(normed[:, :, 123], cmap=plt.cm.bone)

mri = nib.load(os.path.join(base_dir + '/{}/{}'.format(patient, patient + '_t1.nii.gz')))
mri_mat=mri.get_fdata()
m = np.mean(mri_mat)
s = np.std(mri_mat)
plt.imshow(mri_mat[:, :, 123], cmap=plt.cm.bone)

normed=(mri_mat-m)/s


new_img = nib.Nifti1Image(normed, mri.affine, mri.header)
nib.save(new_img, "output.nii.gz")

plt.figure()
plt.title('saved normed')
saved_Img = nib.load("output.nii.gz").get_fdata()
plt.imshow(saved_Img[:, :, 123], cmap=plt.cm.bone)
```
